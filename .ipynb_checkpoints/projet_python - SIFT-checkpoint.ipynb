```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline

path = r'JPG/JPG_GRIS/'
# read images
img1 = cv.imread(path+'clash.jpg')  
img2 = cv.imread(path+'barbare.jpg') 

img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)

#sift
sift = cv.SIFT_create()

keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2,None)

#feature matching
bf = cv.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)
figure = plt.figure(figsize=(16, 9))
plt.title('Keypoints detection with SIFT technique'), plt.xticks([]), plt.yticks([])
plt.imshow(img3),plt.show();
```
